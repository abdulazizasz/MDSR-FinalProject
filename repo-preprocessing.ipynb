Some repositories without a master branch will not be downloaded.

```python
topdir="E:/Data/"
count=0
for repo in repos:
    url=repo['url']+"/archive/master.zip"
    zipfilepath=topdir+"zips/"+repo['reponame']+".zip"
    repodirpath=topdir+"data/"+repo['reponame']+"-master"
    if not os.path.isfile(zipfilepath):
        print(url)
        count+=1
        try:
            urllib.request.urlretrieve(url, zipfilepath)
        except:
            print("Error downloading "+zipfilepath)
            print(sys.exc_info()[0])
            continue
        time.sleep(10.5+random.randrange(20))
        if count%10 == 0: time.sleep(60)
    if not os.path.isdir(repodirpath):
        with zipfile.ZipFile(zipfilepath,"r") as zipf:
            try:
                zipf.extractall(topdir+"data")
            except:
                print("Error unzipping "+zipfilepath)
                print(sys.exc_info()[0])
```

The following name extraction is based on the code from the paper "Topic modeling of public repositories at scale using names in source code", but we are using a slightly different regular expression.

```python
NAME_BREAKUP_RE = re.compile(r"\W+")
def extract_names(token): 
    token = token.strip() 
    prev_p = [""]
    def ret(name): 
        r = name.lower() 
        if len(name) >= 3: 
            yield r 
            if prev_p[0]: 
                yield prev_p[0] + r 
                prev_p[0] = "" 
        else: prev_p[0] = r
        
    for part in NAME_BREAKUP_RE.split(token): 
        if not part: continue 
        prev = part[0] 
        pos = 0 
        for i in range(1, len(part)): 
            this = part[i] 
            if prev.islower() and this.isupper(): 
                yield from ret(part[pos:i]) 
                pos = i 
            elif prev.isupper() and this.islower(): 
                if 0 < i - 1 - pos <= 3: 
                    yield from ret(part[pos:i - 1]) 
                    pos = i - 1 
                elif i - 1 > pos: 
                    yield from ret(part[pos:i]) 
                    pos = i 
            prev = this 
        last = part[pos:] 
        if last: 
            yield from ret(last)
```
